```python
import h5py
import os
import numpy as np
from tqdm.autonotebook import tqdm
import os
```

```python
os.makedirs('datasets_compressed', exist_ok=True)
os.makedirs('datasets_compressed/test_case_I', exist_ok=True)
```

```python
directory = 'datasets/test_case_I'
files = os.listdir(directory)
print(files)

for file in tqdm(files):
    inFile = h5py.File(f'datasets/test_case_I/{file}', 'r')
    outFile = h5py.File(f'datasets_compressed/test_case_I/{file}', 'w')
    outFile.create_group('generatorSettings')
    for k in inFile.attrs.keys():
        outFile.attrs[k] = inFile.attrs[k]
    for k in inFile['generatorSettings'].attrs.keys():
        outFile['generatorSettings'].attrs[k] = inFile['generatorSettings'].attrs[k]
    dataGrp = outFile.create_group('simulationData')
    outFile['simulationData']['dudt'] = inFile['simulationData']['dudt'][:]
    outFile['simulationData']['fluidAreas'] = inFile['simulationData']['fluidAreas'][:]
    outFile['simulationData']['fluidDensity'] = inFile['simulationData']['fluidDensity'][:]
    outFile['simulationData']['fluidPosition'] = inFile['simulationData']['fluidPosition'][:]
    outFile['simulationData']['fluidPressure'] = inFile['simulationData']['fluidPressure'][:]
    outFile['simulationData']['fluidVelocities'] = inFile['simulationData']['fluidVelocities'][:]
    outFile.close()
    inFile.close()
```

```python
os.makedirs('datasets_compressed', exist_ok=True)
os.makedirs('datasets_compressed/test_case_II/test/', exist_ok=True)

directory = 'datasets/test_case_II/test'
files = os.listdir(directory)
# print(files)

for file in tqdm(files):
    inFile = h5py.File(f'datasets/test_case_II/test/{file}', 'r')
    outFile = h5py.File(f'datasets_compressed/test_case_II/test/{file}', 'w')

    # outFile.create_group('generatorSettings')
    for k in inFile.attrs.keys():
        outFile.attrs[k] = inFile.attrs[k]

    # print(inFile['boundaryInformation'].keys(), inFile['boundaryInformation'].attrs.keys())
    outFile.create_group('boundaryInformation')
    for k in inFile['boundaryInformation'].attrs.keys():
        outFile['boundaryInformation'].attrs[k] = inFile['boundaryInformation'].attrs[k]
    for k in ['boundaryPosition', 'boundaryNormals', 'boundaryArea', 'boundaryVelocity']:
        outFile['boundaryInformation'][k] = inFile['boundaryInformation'][k][:]

    simulationData = outFile.create_group('simulationExport')
    for f in tqdm(inFile['simulationExport'].keys(), leave = False):
        data = inFile['simulationExport'][f]
        outGroup = simulationData.create_group(f)
        # print(data.keys())
        if 'fluidGravity' in data.keys():
            newData = {
                'fluidPosition': data['fluidPosition'][:], 
                'fluidVelocity': data['fluidVelocity'][:], 
                'fluidArea': data['fluidArea'][:], 
                'fluidDensity': data['fluidDensity'][:], 
                'fluidSupport': data['fluidSupport'][:], 
                'fluidGravity': data['fluidGravity'][:], 
                'boundaryDensity': data['boundaryDensity'][:]}
        else:
            newData = {
                'fluidPosition': data['fluidPosition'][:], 
                'fluidVelocity': data['fluidVelocity'][:], 
                'fluidArea': data['fluidArea'][:], 
                'fluidDensity': data['fluidDensity'][:], 
                'fluidSupport': data['fluidSupport'][:], 
                'boundaryDensity': data['boundaryDensity'][:]}
        
        for k in newData.keys():
            outGroup.create_dataset(k, data=newData[k])
        for k in data.attrs.keys():
            outGroup.attrs[k] = data.attrs[k]

    outFile.close()
    inFile.keys()
```
